# 7장 - 동시성과 병렬성


- <font color='blue'>동시성</font> 이란, 컴퓨터가 같은 시간에 여러 다른 작업을 처리하는 것처럼 보이는 것을 뜻한다.
- CU 코어가 하나뿐인 컴퓨터는, 여러 프로그램이 번갈아가면서 실행되면서 프로그램이 동시에 수행되는 것 같은 착각을 불러일으킬 수 있다.
<br>
<br>
- <font color='blue'>병렬성</font> 이란, 같은 시간에 여러 다른 작업ㅇ르 실제로 처리하는 것을 뜻한다.
- CPU 코어가 여러 개인 컴퓨터는 여러 프로그램을 동시에 실행할 수 있다.
<br>
<br>
- 병렬성과 동시성의 가장 핵심적인 차이는 <font color='blue'>속도 향상</font> 에 있다.
- 서로 다른 두 실행 경로가 병렬적으로 앞으로 진행되면, 전체 작업 소요 시간은 절반으로 줄어든다.
- 반대로 서로 다른 실행 경로가 있는 동시성 프로그램은 겉으로 볼때는 병렬적으로 실행되는 것처럼 보이지만, 전체 작업 소요시간은 빨라지지 않는다.

- 파이썬을 사용하면 다양한 스타일로 동시성 프로그램을 쉽게 작성할 수 있다.
<br>
<br>
- 스레드는 상대적으로 적은 양의 동시성을 제공하지만
- 코루틴은 수많은 동시성 함수를 사용할 수 있게 해준다.
<br>
<br>
- 파이썬은 시스템 콜(system call), 하위 프로세스(sub process), C 확장(extension) 을 사용해 작업을 병렬로 사용할 수 있다.

sleep.exe 코드

#include "stdafx.h"
#include <stdlib.h>
#include <windows.h>

int main(int argc, char** argv)
{
	DWORD dwPreTime = GetTickCount();

```python
import time
start = time.time()

proc = subprocess.Popen(['sleep', '1'])
while proc.poll() is None:
    print('작업중...')
    pass
    
print('종료 상태', proc.poll())

end = time.time()
delta = end - start
print(f'{delta:.3} 초 만에 끝남')
```

```python
'''
자식 프로세스와 부모를 분리하면 부모 프로세스가 원하는 개수만큼 많은 자식 프로세스를 병렬로 실행할 수 있다.
'''
import time

start = time.time()
sleep_procs = []
for _ in range(10):
    proc = subprocess.Popen(['sleep', '1'])
    sleep_procs.append(proc)
    
for proc in sleep_procs:
    proc.communicate()

end = time.time()
delta = end - start
print(f'{delta:.3} 초 만에 끝남')
```

- 파이썬 프로그램의 데이터를 파이프(pipe)를 사용해, 하위 프로세스로 보내거나, 하위 프로세스의 출력을 받을 수 있다.

```python
import os
def run_encrypt(data):
    env = os.environ.copy()
    env['password'] = 'zf7ShyBhZ0raQDdE/FiZpm/m/8f9X+M1'
    proc = subprocess.Popen(
        ['openssl', 'enc', '-des3', '-pass', 'env:password'],
        env=env,
        stdin=subprocess.PIPE,
        stdout=subprocess.PIPE)
    proc.stdin.write(data)
    proc.stdin.flush() # 자식이 입력을 받도록 보장한다.
    return proc

procs = []
for _ in range(3):
    data = os.urandom(10)
    proc = run_encrypt(data)
    procs.append(proc)
    
for proc in procs:
    out, _ = proc.communicate()
    print(out[-10:])
```

- 자식 프로세스가 결코 끝나지 않는 경우, 입력이나 출력 파이프를 기다리면서 블록되는 경우가 우려된다면 timeout 파라미터를 communicate 메서드에 전달할 수 있다.

```python
proc = subprocess.Popen(['sleep', '10'])
try:
    proc.communicate(timeout=0.1)
except subprocess.TimeoutExpired:
    print('timeout')
    proc.terminate()
    proc.wait()
    
print('종료 상태', proc.poll())
```

### BETTER WAY 53 - 블로킹 I/O의 경우 쓰레드를 사용하고 병렬성을 피하라

- 파이썬의 표준 구현을 <font color='blue'>CPython</font> 이라고 한다.
- CPython 은 두 단계를 거쳐 파이썬 프로그램을 실행한다.
<br><br>
- 첫 번째 단계는, 소스 코드를 구문 분석하여 바이트코드(bytecode)로 변환한다. 바이트코드는 8비트 명령어를 사용하는 저수준 프로그램 표현이다.
- 파이썬 3.6 부터는 16비트 명령어를 사용하기 때문에 워드코드(wordcode)라고 불러야 하지만, 기본적인 아이디어는 똑같다.
<br><br>
- 두 번째 단계는, 이 바이트코드를 스택 기반 인터프리터를 통해 실행한다.
- 바이트코드 인터프리터에는 파이썬 프로그램이 실행되는 동안 일관성있게 유지해야 하는 상태가 존재한다.
- CPython 은 전역 인터프리터 락(Global Interpreter Lock, GIL) 이라는 방법을 사용해 일관성을 강제로 유지한다.

- 근본적으로 GIL 은 상호 배제 락 (Mutex) 이며, CPython 이 선점형(preemptive) 멀티스레드로 인해 영향을 받는 것을 방지한다.
<br><br>
- <font color='red'>GIL 에는 꼭 알아둬야 하는 큰 부작용이 있다.</font>
- C++과 자바와 같이 파이썬도 다중 실행 스레드를 지원하지만, GIL 로 인해 여러 스레드 중 어느 하나만 앞으로 진행할 수 있다.
- 따라서, 파이썬 프로그램의 속도를 높이고 병렬 처리를 수행하고자 쓰레드를 사용한다면 여러분은 크게 실망할 것이다.

```python
# 예) 파이썬으로 계산량이 매우 많은 작업을 수행하고 싶다

def factorize(number):
    for i in range(1, number + 1):
        if number % i == 0:
            yield i
            
import time

numbers = [2139079, 1214759, 1516637, 1852285]
start = time.time()

for number in numbers:
    list(factorize(number))
    
end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```

```python
# 똑같은 계산을 수행하는 파이썬 쓰레드 
from threading import Thread

class FatorizeThread(Thread):
    def __init__(self, number):
        super().__init__()
        self.number = number
        
    def run(self):
        self.factors = list(factorize(self.number))
        
start = time.time()

threads = []
for number in numbers:
    thread = FatorizeThread(number)
    thread.start()
    threads.append(thread)
    
for thread in threads:
    thread.join()
    
end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```

```python
'''
놀랍게도 스레드를 하나만 써서 할 때보다, 시간이 더 오래 걸린다.
사용할 수 있는 CPU 가 많음에도 불구하고, 속도가 느려질 것이라고는 예상하지 못했을 것이다.
이 결과는 표준 CPython 인터프리터에서 프로그램을 사용할 때 GIL 이 미치는 영향을 잘 보여준다.
'''
'''
물론 CPython 에서도 다중 코어를 활용할 수 있는 방법이 있다.
다만 코딩하는데 상당한 노력이 필요하다.
'''

'''
이러한 한계에도 불구하고 파이썬이 스레드를 지원하는 이유는 무엇일까?
첫째, 다중 스레드를 사용하면 프로그램이 동시에 여러 일을 하는 것처럼 보이게 만들기 쉽다
둘째, 블로킹 I/O 를 다루기 위해서다.
'''

# 예) 직렬 포트를 통해 원격 제어 헬리콥터에 신호를 보내고 싶다.
import select
import socket

def slow_systemcall():
    select.select([socket.socket()], [], [], 0.1) # 운영체제에게 0.1 초 블록한 다음에 제어를 돌려달라고 요청
    
start = time.time()

# 이 시스템 콜을 순차적으로 실행하면 실행에 필요한 시간이 선형으로 증가한다.
for _ in range(5):
    slow_systemcall()
    
end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```

```python
# 문제는 slow_systemcall 함수가 실행되는 동안 프로그램이 아무 진전을 이룰 수 없다는 것이다.

# 다음 코드는, slow_systemcall 함수를 여러 쓰레드에서 따로따로 호출한다. 이렇게 하면 여러 직렬 포트와 통신하면서 주 스레드는 필요한 계산을 수행할 수 있다.

start = time.time()

threads = []
for _ in range(5):
    thread = Thread(target=slow_systemcall)
    thread.start()
    threads.append(thread)
    
for thread in threads:
    thread.join()
    
end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```

- 이런 GIL 로 인해 생기는 한계가 있더라도, 파이썬이 여러 스레드를 통해 시스템 콜을 병렬로 실행할 수 있음을 보여준다.

### BETTER WAY 54 - 스레드에서 데이터 경합을 피하기 위해 Lock 을 사용하라

- GIL 이 동시 접근을 보장해주는 락 역할을 하는 것처럼 보이지만, 실제로는 그렇지 않다.

```python
class Counter:
    def __init__(self):
        self.count = 0
        
    def increment(self, offset):
        self.count += offset
        
# 센서를 읽을 때는 블로킹 I/O를 수행하므로 센서마다 작업자 스레드를 할당한다고 하자.
def worker(sensor_index, how_many, counter):
    for _ in range(how_many):
        # 센서를 읽는다.
        #...
        counter.increment(1)

# 다음 코드는 병렬로 센서마다 하나씩 worker 스레드를 실행하고, 모든 스레드가 값을 다 읽을 때까지 기다린다.
from threading import Thread

how_many = 10**5
counter = Counter()

import time
start = time.time()

threads = []
for i in range(5):
    thread = Thread(target=worker, args=(i, how_many, counter))
    threads.append(thread)
    thread.start()
    
for thread in threads:
    thread.join()
    
expected = how_many * 5
found = counter.count
print(f'카운터 값은 {expected} 이어야 하는데, 실제로는 {found} 입니다.')

end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```

```python
'''
실제 실행한 결과는 예상과 전혀 다르다!
파이썬 인터프리터는 실행되는 모든 쓰레드를 강제로 공평하게 취급해서 각 스레드의 실행 시간을 거의 비슷하게 만든다.
이를 위해 파이썬은 실행 중인 스레드를 일시 중단 시키고, 다른 스레드를 실행시키는 일을 반복한다.
문제는 파이썬이 스레드를 언제 중단시킬지 알 수 없기 때문에,
연산을 수행하는 도중에 파이썬이 쓰레드를 일시 중단 시킬 수 있다.
이로 인해 위 같은 결과가 생긴다.

간단해 보이는 self.count += offset 구문도
실제로는
temp = get(self.count)
value = temp + offset
set(self.count, value)
이런식으로 세가지 연산으로 이뤄지는데, 이 중간 과정에서 스레드가 일시 중단될 수 있다는 것이다.

이런 일시 중단으로 인해 스레드간 연산 순서가 뒤섞이면서 값이 꼬일 수 있는 것이다.
'''

'''
이와 같은 데이터 경합이나 다른 유형의 데이터 구조 오염을 해결하기 위해 파이썬은 threading 내장 모듈 안에 여러 가지 튼튼한 도구를 제공한다.
가장 간단하지만 유용한 도구로 Lock 클래스가 있다.
Lock 클래스는 상호 배제 락(뮤텍스) 이다.
'''

from threading import Lock

class LockingCounter:
    def __init__(self):
        self.lock = Lock()
        self.count = 0
        
    def increment(self, offset):
        with self.lock:
            self.count += offset
            
from threading import Thread

how_many = 10**5
counter = LockingCounter()

import time
start = time.time()

threads = []
for i in range(5):
    thread = Thread(target=worker, args=(i, how_many, counter))
    threads.append(thread)
    thread.start()
    
for thread in threads:
    thread.join()
    
expected = how_many * 5
found = counter.count
print(f'카운터 값은 {expected} 이어야 하는데, 실제로는 {found} 입니다.')

end = time.time()
delta = end - start
print(f'총 {delta:.3f} 초 걸림')
```
